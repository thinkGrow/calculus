```python
#Integration using tapezoidal rule, for y = x, in limit 0 to 2
import numpy as np 
import matplotlib.pyplot as plt
a = 0.0
b = 2.0

N = int(input("Enter your value for N: "))

#N = 1000; #N is the number of points

h = (b-a)/N

x = np.linspace(a,b,N)
y = x; #this is the given function y=f(x)

f = 0.0
for k in range(len(x)-1):
    f+=0.5 * ((x[k+1]-x[k]) * (y[k+1]+y[k]) )
print(f)

plt.plot(x, y)
plt.show()
```
